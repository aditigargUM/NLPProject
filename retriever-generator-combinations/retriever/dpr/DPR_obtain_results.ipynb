Obtain the top 10 passages using facebook multiset embeddings for queries and passages

```python
import faiss
import pickle
import numpy as np

def obtain_and_save_top10(passages_faiss_index_filepath, queries_embeddings_filepath, results_filepath):
  
  passages_faiss_index = faiss.read_index(passages_faiss_index_filepath)

  with open(queries_embeddings_filepath, "rb") as fp:   # Unpickling
      queries_embeddings = pickle.load(fp)

  queries_embeddings = np.stack(queries_embeddings)
  D, I = passages_faiss_index.search(queries_embeddings, 10)

  with open(results_filepath, "wb") as fp:   # Unpickling
    pickle.dump(I, fp)
```
